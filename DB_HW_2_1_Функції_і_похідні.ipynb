1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Визначаємо символічну змінну
x = sp.symbols('x')

# Лінійна функція
f = 10 * x - 5
df = sp.diff(f, x)

print("Похідна лінійної функції 10x-5:", df)
```

```python
x = np.linspace(0, 10, 100)

f = 10 * x - 5
df = np.full_like(x, 10)

plt.figure(figsize=(10, 5))

plt.plot(x, f, label="f(x) = 10x - 5", color="blue", linestyle='solid')
plt.plot(x, df, label="f'(x) = 10", color="green", linestyle='dashdot')

plt.title("Графік функції f(x) = 10x - 5 та її похідної")
plt.xlabel("Кількість годин x")
plt.ylabel("Оцінка f(x)")
plt.legend()


plt.grid(True)
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

# Квадратична функція
f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)

# Експоненційна функція
f_ln = sp.ln(x)
derivative_f_ln = sp.diff(f_ln, x)

print("Похідна квадратичної функції sin(x):", derivative_f_sin)
print("Похідна експоненційної функції ln(x):", derivative_f_ln)
```

```python
xs = sp.symbols('x')


f_sin = sp.sin(xs)
derivative_f_sin = sp.diff(f_sin, xs)

#перетворення з NumPy
f_sin_np = sp.lambdify(xs, f_sin, 'numpy')
derivative_f_sin_np = sp.lambdify(xs, derivative_f_sin, 'numpy')

x = np.linspace(0, 10, 100)

f = f_sin_np(x)
df = derivative_f_sin_np(x)


plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = sin(x)', color='blue')
plt.plot(x, df, label="f'(x) = cos(x)", color='green')

plt.title("Графік функції sin(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
xs = sp.symbols('x')


f_ln = sp.ln(xs)
derivative_f_ln = sp.diff(f_ln, xs)

#перетворення з NumPy
f_ln_np = sp.lambdify(xs, f_ln, 'numpy')
derivative_f_ln_np = sp.lambdify(xs, derivative_f_ln, 'numpy')

x = np.linspace(0.1, 20, 50)

f = f_ln_np(x)
df = derivative_f_ln_np(x)


plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = ln(x)', color='blue')
plt.plot(x, df, label="f'(x) = 1/x", color='green')

plt.title("Графік функції ln(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')


# функція та похідна
f_sq = x**2 - 5 * x + 1
derivative_f_sq = sp.diff(f_sq, x)

print("Похідна квадратичної функції x**2−5*x+1:", derivative_f_sq)

#перетворення з NumPy
f_sq_np = sp.lambdify(xs, f_sq, 'numpy')
derivative_f_sq_np = sp.lambdify(xs, derivative_f_sq, 'numpy')

x = np.linspace(0, 100, 100)

f = f_sq_np(x)
df = derivative_f_sq_np(x)


plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = x**2−5*x+1', color='blue')
plt.plot(x, df, label="f'(x) = 2*x - 5", color='green')

plt.title("Графік функції x**2−5*x+1 та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
